#  Average Rating by Genre: 

```python
df_rate_bar = df1.groupby('Genre').agg({'IMDB_Rating':'mean'}).reset_index().sort_values('IMDB_Rating')
df_rate_bar.head()
```

```python
rate_bar = sns.barplot(x='Genre',y='IMDB_Rating',data=df1,order=df_rate_bar['Genre'])
rate_bar.set(title ='Average Rating by Genre')
plt.xticks(rotation=90)
```

# Top Movies by Genre

```python
df_top_movies=df1.groupby('Genre').agg({'IMDB_Rating':'max'}).reset_index()
```

```python
df_top_movies=df_top_movies.merge(df1,on=['Genre',"IMDB_Rating"]).drop_duplicates('Genre').sort_values('IMDB_Rating')
df_top_movies['Title']=df_top_movies['Genre'].astype(str)+'-'+df_top_movies['Series_Title']
df_top_movies.head()
```

```python
bar=sns.barplot(data=df_top_movies,x='Genre',y='IMDB_Rating',order=df_top_movies['Genre'])
bar.set_xticklabels(df_top_movies['Title'],rotation=90)
bar.set(xlabel ="Genre - Movie", title ='Top Movies by Genre')
plt.show()
```
